```python

import math

numerical_cols = df.select_dtypes(include=['number']).columns

# Calculate number of rows and columns for subplots
num_cols = 4  # Number of columns
num_rows = 3  # Calculate rows

# Create a single figure with subplots in a grid
fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, 10))

# Iterate through numerical columns and plot on subplots
for i, col in enumerate(numerical_cols):
    row = i // num_cols  # Calculate subplot row index
    col_index = i % num_cols  # Calculate subplot column index
    
    sns.histplot(df[col], ax=axes[row, col_index])
    axes[row, col_index].set_title(f'Histogram of {col}')
    axes[row, col_index].set_xlabel(col)
    axes[row, col_index].set_ylabel('Frequency')

# Adjust layout to prevent overlapping
plt.tight_layout()
plt.show()
```
