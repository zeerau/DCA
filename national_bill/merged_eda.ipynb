```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option("display.max_rows", None)
```

```python
df = pd.read_csv(r'C:\Users\DELL\Documents\DCA\national_bill\2025_budget_merged.csv')

df.sample(100)
```

```python
df['LINE ITEM'].value_counts(ascending=False)
```

```python
motor_cost = df[df['LINE ITEM'] == 'PURCHASE OF MOTOR VEHICLES']
motor_cost_by_ministry = motor_cost.groupby(by='MINISTRY')['AMOUNT'].agg('sum').sort_values(ascending=False)[:10]
motor_cost_by_ministry

sns.barplot(x=motor_cost_by_ministry.values,y=motor_cost_by_ministry.index)
plt.title('PURCHASE OF MOTOR VEHICLES')
plt.ticklabel_format(axis='x', style='plain')
plt.xticks(rotation=90)
plt.show()
```

```python
motor_cost = df[df['LINE ITEM'] == 'SUBSCRIPTION TO PROFESSIONAL BODIES']
motor_cost_by_ministry = motor_cost.groupby(by='MINISTRY')['AMOUNT'].agg('sum').sort_values(ascending=False)[:10]
motor_cost_by_ministry

sns.barplot(y=motor_cost_by_ministry.index,x=motor_cost_by_ministry.values,orient='h')
plt.title('SUBSCRIPTION TO PROFESSIONAL BODIES')
plt.ticklabel_format(axis='x', style='plain')
plt.xticks(rotation=90)
plt.show()
```
